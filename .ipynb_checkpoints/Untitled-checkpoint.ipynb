```python
import os
import numpy as np
import pandas as pd
import GetOldTweets3 as got
import itertools
import collections
import re
import json
from datetime import datetime, timedelta
import time
```

```python
# Function to remove urls from tweets


def remove_url(txt):

    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())


# Number of tweets to retrieve
count = 50

# List of tweets retrieved
tweet_list = []


# Dictionary to which the tweets are added
def tweet_to_dict(tweet):
    tweet_dict = {
        'user': tweet.username,
        'id': tweet.id,
        ## Fixed time from UTC to UTC+3
        'date': str(tweet.date+timedelta(hours=3)),
        'hashtags': tweet.hashtags,
        'text': remove_url(tweet.text),
    }
    return tweet_dict


# Adding tweets to dictionary and then to list
def add_to_list():
    for tweet in tweets:
        tweet_dict_temp = tweet_to_dict(tweet)
        tweet_list.append(tweet_dict_temp)
    print('--- Added to list!--- ')
```

```python
# Date to start from
dateSince = datetime(2020, 8, 1)
dateUntil = datetime(2020, 9, 1)
num_of_days = (dateUntil - dateSince).days
# Converting dates to strings
strDateSince = dateSince.strftime("%Y-%m-%d")
strDateUntil = dateUntil.strftime("%Y-%m-%d")
print(num_of_days)
```

```python
# Cycling through days
for day in range(num_of_days):
    # Create a custom search term and define the number of tweets
    print("--- Grabbing " + strDateSince + " tweets... ---")
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(
        'Coronavirus').setSince(strDateSince).setUntil(strDateUntil).setLang('en').setMaxTweets(count)
    # Call getTweets and saving in tweets
    print('--- Starting query... ---')
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    print('--- Adding to list... ---')
    add_to_list()
    print('--- Writing JSON... ---')
    # Saving list to JSON file
    with open('saver_output1.json', 'w') as stream:
        json.dump(tweet_list,stream)
    print('Total number of tweets: ', len(tweet_list))
    print('--- Going to sleep... ---')
    time.sleep(60*5)
    print('--- Changing day ... ---\n\n')
    dateSince += timedelta(days=1)
    dateUntil += timedelta(days=1)
```
